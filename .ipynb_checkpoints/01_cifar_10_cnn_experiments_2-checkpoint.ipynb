#### Lab 0 - 0.1 - CNN Experiments on CIFAR-10

##### Baseline: SGD (lr=0.0001) | LeakyReLU
##### Experiment 1: Adam | LeakyReLU
##### Experiment 2: Adam | Tanh

```python
%%time

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.tensorboard import SummaryWriter
import matplotlib.pyplot as plt
import numpy as np
import random

# Reproducibility
torch.manual_seed(42)
torch.cuda.manual_seed_all(42)
np.random.seed(42)
random.seed(42)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# 25% GPU
torch.cuda.set_per_process_memory_fraction(0.25, device=0)

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)
```

---
### Loading & Preparing the CIFAR-10 Dataset

**CIFAR-10** is a widely-used image classification benchmark containing **60,000 color images** (32×32 pixels, 3 RGB channels) across **10 classes**: airplane, car, bird, cat, deer, dog, frog, horse, ship, truck. It is split into 50,000 training and 10,000 test images.

**Transforms — what happens to each image before it enters the network:**

1. **`ToTensor()`** — Converts a PIL image (values 0–255, shape H×W×C) into a PyTorch tensor (values 0.0–1.0, shape C×H×W). PyTorch convolutions expect the *channels-first* format.

2. **`Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))`** — For each channel: `pixel = (pixel - 0.5) / 0.5`, which rescales values from [0, 1] to [−1, +1]. Centering data around zero helps gradient-based optimization — imagine trying to balance a seesaw that starts tilted vs. one that starts level.

**DataLoader choices:**
- **`batch_size=64`** — We feed 64 images at a time (a "mini-batch"). This balances between noisy single-sample gradients and the expensive full-dataset gradient.
- **`shuffle=True` (train only)** — Randomize sample order each epoch so the model doesn't memorize sequences.
- **`num_workers=2`** — Two background processes load the next batch while the GPU trains on the current one.

```python
%%time

# Transforms
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Download and load training set
trainset = torchvision.datasets.CIFAR10(
    root='./data', train=True, download=True, transform=transform
)
trainloader = torch.utils.data.DataLoader(
    trainset, batch_size=64, shuffle=True, num_workers=2
)

# Download and load test set
testset = torchvision.datasets.CIFAR10(
    root='./data', train=False, download=True, transform=transform
)
testloader = torch.utils.data.DataLoader(
    testset, batch_size=64, shuffle=False, num_workers=2
)

# CIFAR-10 class labels
classes = ('plane', 'car', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck')

print('Training samples:', len(trainset))
print('Test samples:', len(testset))
```

#### The CNN Architecture

**Convolutional Block (feature extractor)** : three stages of Conv → Activate → MaxPool:

| Stage | Operation | Shape Change | Intuition |
|-------|-----------|-------------|----------|
| 1 | Conv2d(3→32, 3×3, pad=1) → Activation → MaxPool(2×2) | 3×32×32 → 32×16×16 | Learns 32 low-level filters (edges, color blobs). `padding=1` preserves spatial size before pooling. MaxPool halves the resolution, keeping only the strongest signal in each 2×2 window. |
| 2 | Conv2d(32→64, 3×3, pad=1) → Activation → MaxPool(2×2) | 32×16×16 → 64×8×8 | 64 mid-level filters combine edges into textures and shapes. |
| 3 | Conv2d(64→128, 3×3, pad=1) → Activation → MaxPool(2×2) | 64×8×8 → 128×4×4 | 128 high-level filters detect object parts. |

**Fully Connected Block (classifier):**

| Layer | Shape Change | Purpose |
|-------|-------------|--------|
| Flatten | 128×4×4 → 2048 | Reshape 3D feature maps into a 1D vector. |
| Linear(2048→256) + Activation | 2048 → 256 | Learn which feature combinations matter for classification. |
| Linear(256→10) | 256 → 10 | Output one score (logit) per class. Highest score = predicted class. |

#### The Training Loop (`train_model`)

For each epoch, every mini-batch goes through:
1. **Forward pass** - images flow through the CNN to produce 10 class scores.
2. **Loss computation** - `CrossEntropyLoss` measures how far the predicted distribution is from the true label. It internally applies Softmax + Negative Log-Likelihood, so we output raw logits.
3. **Backward pass** (`loss.backward()`) - computes the gradient of the loss w.r.t. every weight in the network via backpropagation.
4. **Parameter update** (`optimizer.step()`) - adjusts weights using the chosen optimizer.
5. **`optimizer.zero_grad()`** - resets gradients to zero before the next batch, because PyTorch *accumulates* gradients by default.

#### The Evaluation Function (`evaluate_model`)

- Switches the model to **eval mode** (`model.eval()`) - disables dropout and uses running stats for batch norm.
- Wraps inference in **`torch.no_grad()`** - tells PyTorch not to track gradients, saving memory and computation since we're only reading predictions, not training.
- Computes **test accuracy** = (correctly classified images / total images) × 100.

```python
%%time

class CNN(nn.Module):
    def __init__(self, activation_fn=nn.LeakyReLU(negative_slope=0.01)):
        super(CNN, self).__init__()
        self.activation_fn = activation_fn

        # Convolutional layers
        self.conv_block = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            self.activation_fn,
            nn.MaxPool2d(2, 2),                          

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            self.activation_fn,
            nn.MaxPool2d(2, 2),                          

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            self.activation_fn,
            nn.MaxPool2d(2, 2)                           
        )

        # Fully connected layers
        self.fc_block = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 4 * 4, 256),
            self.activation_fn,
            nn.Linear(256, 10)
        )

    def forward(self, x):
        x = self.conv_block(x)
        x = self.fc_block(x)
        return x


def train_model(model, optimizer, num_epochs, run_name):
    """Training function."""
    criterion = nn.CrossEntropyLoss()
    writer = SummaryWriter(f'runs/{run_name}')

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for i, (inputs, labels) in enumerate(trainloader):
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

        epoch_loss = running_loss / len(trainloader)
        epoch_acc = 100. * correct / total

        # Tensorboard logging
        writer.add_scalar('Loss/train', epoch_loss, epoch)
        writer.add_scalar('Accuracy/train', epoch_acc, epoch)

        print(f'Epoch [{epoch+1}/{num_epochs}] | Loss: {epoch_loss:.4f} | Train Acc: {epoch_acc:.2f}%')

    writer.close()
    print('Training complete!')
    return model


def evaluate_model(model, label):
    """Evaluation function."""
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in testloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    accuracy = 100. * correct / total
    print(f'Test Accuracy ({label}): {accuracy:.2f}%')
    return accuracy


# Final summary
results = {}
NUM_EPOCHS = 10
```

### Experiment 0 : Baseline: SGD (lr=0.0001) + LeakyReLU

**Optimizer - SGD (Stochastic Gradient Descent):**  
The simplest optimizer. It updates each weight by subtracting `lr × gradient` with `lr=0.0001`.

**Activation - LeakyReLU (negative_slope=0.01):**  
LeakyReLU outputs `x` if `x > 0`, and `0.01 × x` if `x ≤ 0`. Unlike standard ReLU (which outputs exactly 0 for negative inputs), the small negative slope keeps a tiny gradient flowing even for negative values. This prevents "dead neurons" - neurons that stop learning because their gradient is always zero.

**Expected result:** Very poor accuracy (~10%, i.e., random guessing on 10 classes). The learning rate is too small for SGD to make meaningful progress in 10 epochs.

```python
%%time

# Instantiate model with LeakyReLU
model_exp0 = CNN(activation_fn=nn.LeakyReLU(negative_slope=0.01)).to(device)

# SGD optimizer with lr=0.0001
optimizer_exp0 = optim.SGD(model_exp0.parameters(), lr=0.0001)

print('Optimizer: SGD | LR: 0.0001 | Activation: LeakyReLU (slope=0.01)')

model_exp0 = train_model(model_exp0, optimizer_exp0, NUM_EPOCHS, 'exp0_sgd_leakyrelu')
```

```python
%%time

results['Experiment 0 (SGD + LeakyReLU)'] = evaluate_model(model_exp0, 'Experiment 0 - SGD + LeakyReLU')
```

### Experiment 1 - Adam (lr=0.001) + LeakyReLU

**Adam (Adaptive Moment Estimation):**  
Maintains **two running averages** for each weight:
- **First moment (mean of gradients)** - acts like *momentum*, smoothing out noisy gradient oscillations.
- **Second moment (mean of squared gradients)** - acts like a *per-parameter learning rate*. Parameters with consistently large gradients get a smaller effective step; parameters with small gradients get a larger step. This is the "adaptive" part.

Adam also applies bias correlation in early training steps, because the running averages start at zero and would otherwise underestimate the true moments.

**Expected result:** Significant improvement - the model should reach ~75% test accuracy, showing that optimizer choice is one of the most impactful hyperparameter decisions.

```python
%%time

# Instantiate model with LeakyReLU
model_exp1 = CNN(activation_fn=nn.LeakyReLU()).to(device)

# Adam optimizer (lr=0.001)
optimizer_exp1 = optim.Adam(model_exp1.parameters(), lr=0.001)

print('Optimizer: Adam | LR: 0.001 | Activation: LeakyReLU')

model_exp1 = train_model(model_exp1, optimizer_exp1, NUM_EPOCHS, 'exp1_adam_leakyrelu')
```

```python
%%time

results['Experiment 1 (Adam + LeakyReLU)'] = evaluate_model(model_exp1, 'Experiment 1 - Adam + LeakyReLU')
```

### Experiment 2 — Adam (lr=0.001) + Tanh

**Tanh:**
Tanh squashes any input into the range (−1, +1) using the formula tanh(x) = (eˣ − e⁻ˣ) / (eˣ + e⁻ˣ). It is zero-centered (which helps optimization), but suffers from the vanishing gradient problem

**Vanishing Gradient issue:** When Tanh inputs become large (positive or negative), the output flattens to ±1 and the gradient nearly vanishes. During backpropagation, these tiny gradients get multiplied together across layers, making learning slower for deep networks. LeakyReLU avoids this because its gradient never goes to zero.

**Expected result:** Slightly lower accuracy than Experiment 1 (~72% vs ~75%). Both use Adam, so the optimizer is equally effective. The performance gap comes from Tanh's gradient saturation slowing down learning compared to LeakyReLU, especially in the deeper layers.

```python
%%time

# Instantiate model with Tanh
model_exp2 = CNN(activation_fn=nn.Tanh()).to(device)

# Adam optimizer (lr=0.001)
optimizer_exp2 = optim.Adam(model_exp2.parameters(), lr=0.001)

print('Optimizer: Adam | LR: 0.001 | Activation: Tanh')

model_exp2 = train_model(model_exp2, optimizer_exp2, NUM_EPOCHS, 'exp2_adam_tanh')
```

```python
%%time

results['Experiment 2 (Adam + Tanh)'] = evaluate_model(model_exp2, 'Experiment 2 - Adam + Tanh')
```

### Results

- **Exp 0 → Exp 1 (SGD → Adam):** The jump from ~10% to ~76% shows that the optimizer (and its learning rate) is the dominant factor here. SGD with a tiny lr simply couldn't navigate the loss landscape fast enough.
- **Exp 1 → Exp 2 (LeakyReLU → Tanh):** The smaller drop from ~76% to ~73% shows the activation function matters too - LeakyReLU's non-saturating gradients give it an edge over Tanh, especially as networks get deeper.

```python
print(f'{"Experiment":<35} {"Test Accuracy":>15}')
for name, acc in results.items():
    print(f'{name:<35} {acc:>14.2f}%')
```
